```python
def grab_file(local_path,fname,url_base,url_dir):
    if not os.path.isfile(local_path+fname):
        response = get(url_base+url_dir+fname)
        if response.status_code == 200:
            print('Found file for '+type_year_code+'. Downloading to data/SWDB/'+fname)
            with open(local_path+fname,'wb') as f:
                f.write(response.content)
    else:
        print('Already have '+fname+' locally. Skipping download')
```

```python
#'http://statewidedatabase.org/pub/data/G16/state/state_g16_sov_data_by_g16_srprec.zip'
#'http://statewidedatabase.org/pub/data/G16/state/state_g16_registration_by_g16_srprec.zip'
#'http://statewidedatabase.org/pub/data/G16/state/state_g16_voters_by_g16_srprec.zip'
#'http://statewidedatabase.org/pub/data/G16/state/srprec_state_g16_v01_shp.zip'
#year_str = '16'
#type_code = 'g'
year_strs = [str(x)[-2:] for x in range(2002,2017)]
type_codes = ['g','p','s']
type_year_codes = []
local_path = 'data/SWDB/'
url_base = 'http://statewidedatabase.org/pub/data/'

for year_str in year_strs:
    for type_code in type_codes:
        print('\n')
        url_dir = type_code.upper()+year_str+'/state/'
        type_year_code = type_code+year_str
        if year_str == '03':
            url_dir = 'RECALL/state/'
        
        # Get the Statement of Vote, Statement of Registration, SOR for voters, shapefiles (try a couple filenames)
        fnames = ['state_'+type_year_code+'_sov_data_by_'+type_year_code+'_srprec.zip',
                  'state_'+type_year_code+'_registration_by_'+type_year_code+'_srprec.zip',
                  'state_'+type_year_code+'_voters_by_'+type_year_code+'_srprec.zip',
                  'srprec_state_'+type_year_code+'_v01_shp.zip',
                  'srprec_state_'+type_year_code+'_shp.zip',
                  'srprec_'+type_year_code+'_state_shp.zip'
                ]
        for fname in fnames:
            grab_file(local_path,fname,url_base,url_dir)
```
